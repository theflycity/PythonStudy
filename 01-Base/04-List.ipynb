定义
列表是一种用于储存数据的数据结构

基本使用

```python
import random

counters = [0] * 6
# 模拟掷色子记录每种点数出现的次数
for _ in range(6000):
    face = random.randrange(1, 7)
    counters[face - 1] += 1
# 输出每种点数出现的次数
for face in range(1, 7):
    print(f'{face}点出现了{counters[face - 1]}次')
```

增-删

```python
import random

language_list = ['java', 'python', 'c', 'c++', 'c#', 'php']
print(language_list)
# 在末尾增加元素
language_list.append('go')
print('append-list:',language_list)
# 在指定位置增加元素
language_list.insert(2, 'javascript')
print('insert-list:{}'.format(language_list))
language_list.remove('c#')
print(f'remove-list:{language_list}')
language_list.pop(2)
print('clear-list'+str(language_list))
```
