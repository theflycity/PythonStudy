in 后面的是一个可迭代对象，可以是函数，迭代器，字符串，元组，列表，字典，集合

```python
from boltons.strutils import format_int_list

total = 0
for i in range(1, 101):
    if i % 2 == 0:
        total += i
print(total)
```

02-while

```python
total = 0
i = 1
while i <= 100:
    total += i
    i += 1
print(total)
```

03-break&continue

```python
total = 0
i = 2
while True:
    total += i
    i += 2
    if i > 100:
        break
print(total)
```

04-100以内的素数

```python
for a in range(2, 100):
    mark = True
    for b in range(2, int(a ** 0.5) + 1):
        if a % b == 0:
            mark = False
            break
    if mark:
        print(a)
```

06-花旗骰

```python
import random

money = 1000
while money > 0:
    print(f'你的总资产为: {money}元')
    # 下注金额必须大于0且小于等于玩家的总资产
    while True:
        debt = int(input('请下注: '))
        if 0 < debt <= money:
            break
    # 用两个1到6均匀分布的随机数相加模拟摇两颗色子得到的点数
    first_point = random.randrange(1, 7) + random.randrange(1, 7)
    print(f'\n玩家摇出了{first_point}点')
    if first_point == 7 or first_point == 11:
        print('玩家胜!\n')
        money += debt
    elif first_point == 2 or first_point == 3 or first_point == 12:
        print('庄家胜!\n')
        money -= debt
    else:
        # 如果第一次摇色子没有分出胜负，玩家需要重新摇色子
        while True:
            current_point = random.randrange(1, 7) + random.randrange(1, 7)
            print(f'玩家摇出了{current_point}点')
            if current_point == 7:
                print('庄家胜!\n')
                money -= debt
                break
            elif current_point == first_point:
                print('玩家胜!\n')
                money += debt
                break
print('你破产了, 游戏结束!')
```

```python
import random
def play_game(money):
    debt = int(input('请下注: '))
    first_point = random.randrange(1, 7) + random.randrange(1, 7)
    print(f'\n玩家摇出了{first_point}点')
    if first_point == 7 or first_point == 11:
        print('玩家胜!\n')
        money += debt
        print(f'你的总资产为: {money}元')
        return money
    elif first_point == 2 or first_point == 3 or first_point == 12:
        print('庄家胜!\n')
        money -= debt
        print(f'你的总资产为: {money}元')
        return money
    else:
        # 如果第一次摇色子没有分出胜负，玩家需要重新摇色子
        while True:
            current_point = random.randrange(1, 7) + random.randrange(1, 7)
            print(f'玩家摇出了{current_point}点')
            if current_point == 7:
                print('庄家胜!\n')
                money -= debt
                print(f'你的总资产为: {money}元')
                return money
            elif current_point == first_point:
                print('玩家胜!\n')
                money += debt
                print(f'你的总资产为: {money}元')
                return money
money = 1000
while money > 0:
    money = play_game(money)
print('你破产了, 游戏结束!')


```
