格式化-占位符

```python
# %
a = 321
b = 123
print('%d * %d = %d' % (a, b, a * b))
```

```python
# str.format
a = 321
b = 123
print('{0} * {1} = {2}'.format(a, b, a * b))
```

```python
# f {}
a = 321
b = 123
print(f'{a} * {b} = {a * b}')
```
